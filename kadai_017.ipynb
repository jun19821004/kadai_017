```python
class Human():
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は大人です。")
    else:
      print(f"{self.name}は大人ではありません。")
```

```python
person1 = Human("侍太郎",42)
person2 = Human("侍花子",18)
person3 = Human("侍二郎",12)
person4 = Human("侍三郎",30)
```

```python
people = [person1,person2,person3,person4]
for all_person in people:
  all_person.check_adult()
```
